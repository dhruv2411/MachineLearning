A.	Develop a decision tree classification software on your own and apply it to develop a classification model for Donor dataset 

```python
import pandas as pd
import os
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import LeaveOneOut
from sklearn.model_selection import cross_val_score

df = pd.read_csv("Donor.data",names=['Name','Age','Salary','Donor'])
cols=['Age','Salary']
X = df[cols] # Features
y = df.Donor # Target variable

clf = DecisionTreeClassifier(criterion = 'entropy',random_state = 50, max_depth=2)
#clf.fit(X, y)
clf = clf.fit(X,y)

#Predict the response for test dataset
y_pred = clf.predict(X)

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y, y_pred))
fig = plt.figure(figsize=(5,5))
_ = tree.plot_tree(clf, feature_names=['Age','Salary'], class_names=['N','Y'], filled=True)
model=DecisionTreeClassifier()

leave_validation=LeaveOneOut()

results=cross_val_score(model,X,y,cv=leave_validation)


#print(results)
#print()
print(results.mean())
```

#### Findings:
Here 11 sample are taken and 6 sample satisfies the condition of salary and 5 sample doesn’t and divided into class Y and N.
The accuracy of the data is found to be 1.
The Mean is 0.63
