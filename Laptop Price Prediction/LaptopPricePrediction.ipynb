# Data Cleaning

```python
df.drop(['laptop_ID'],axis=1,inplace=True)
```

```python
df['Company'].value_counts()
```

```python
df['Product'].value_counts()
```

```python
df["Product"] = df["Product"].str.split("(").apply(lambda x: x[0])
```

```python
df['TypeName'].value_counts()
```

```python
df['ScreenResolution'].value_counts()
```

Converting ScreenResolution into different columns

```python
df["IPS"] = df["ScreenResolution"].apply(lambda x: 1 if "IPS" in x else 0 )
df["Touchscreen"] = df["ScreenResolution"].apply(lambda x: 1 if "Touchscreen" in x else 0 )
```

Converting CPU into different columns

```python
df['Cpu'].value_counts()
```

```python
df["CPU Speed"] = df["Cpu"].str.split(" ").apply(lambda x: x[-1]).str.replace("GHz", "")
df["CPU Vendor"] = df["Cpu"].str.split(" ").apply(lambda x: x[0])
df["CPU Type"] = df["Cpu"].str.split(" ").apply(lambda x: x[1:4] if x[1] == "Celeron" and "Pentium" and "Xeon" else (x[1:3] if (x[1] == "Core" or x[0] == "AMD") else x[0]) )
df["CPU Type"] = df["CPU Type"].apply(lambda x: ' '.join(x)) #Joining list
df.drop(['Cpu'],axis=1,inplace=True)
```

```python
df['Ram'].value_counts()
```

```python
df["Ram"] = df["Ram"].str.replace("GB", "").astype('int')
```

```python
df['Memory'].unique()
```

```python
#Split using + first 
#Then Normal split on both
df[['Primary','Secondary']] = df.Memory.str.split("+",n=1,expand=True)
```

```python
df['Primary'].unique()
```

```python
df['Secondary'].unique()
```

```python
df[['Primary Storage','P Storage Type']] = df.Primary.str.split(n=1,expand=True)
df[['Secondary Storage','S Storage Type']] = df.Secondary.str.split(n=1,expand=True)
```

```python
df['P Storage Type'].unique()
```

```python
df["P Storage Type"] = df["P Storage Type"].str.replace("SSD ", "SSD")
df["P Storage Type"] = df["P Storage Type"].str.replace("HDD ", "HDD")
df["P Storage Type"] = df["P Storage Type"].str.replace("Flash Storage ", "Flash Storage")
```

```python
df['S Storage Type'].unique()
```

```python
df['Primary Storage'].unique()
```

```python
df['Secondary Storage'].unique()
```

```python
df.drop(['Memory','Primary','Secondary'],axis=1,inplace=True)
```

```python
def storage(x):
    if x == '1TB' or x == '1.0TB':
        return '1024GB'
    elif x == '2TB':
        return "2048GB"
    else:
        return x

df['Primary Storage'] = df['Primary Storage'].apply(storage)
df['Secondary Storage'] = df['Secondary Storage'].apply(storage)
```

```python
df["Primary Storage"] = df["Primary Storage"].str.replace("GB", "").astype('float')
df["Secondary Storage"] = df["Secondary Storage"].str.replace("GB", "").astype('float')
```

```python
df[['GPU Vendor','Vendor Type']] = df.Gpu.str.split(n=1,expand=True)
df.drop(['Gpu'],axis=1,inplace=True)
```

```python
df['OpSys'].unique()
```

```python
def os(x):
    if x == 'Windows 10' or x == 'Windows 7' or x == 'Windows 10 S':
        return 'Windows'
    elif x == 'macOS' or x == 'Mac OS X':
        return 'Mac'
    elif x == 'Linux':
        return 'Linux'
    elif x == 'Chrome OS':
        return 'Chrome'
    elif x == 'Android':
        return 'Android'
    else:
        return 'No OS'

df['OS'] = df['OpSys'].apply(os)
```

```python
df.drop(['OpSys'],axis=1,inplace=True)
```

```python
df["Weight"] = df["Weight"].str.replace("kg", "").astype('float')
```

# Feature Engineering

```python
df1=df.copy()
df1.info()
```

```python
df1['Company']=pd.factorize(df1['Company'])[0]
df1['TypeName']=pd.factorize(df1['TypeName'])[0]
```

```python
df1=df1.drop(['Product'],axis=1,)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
inches=np.array(df1['Inches'])
inches=inches.reshape(-1,1)
df1['Inches'] = scaler.fit_transform(inches)
```

```python
df1["SRSize"] = df["ScreenResolution"].str.split(" ").apply(lambda x: x[-1])
df1["HD Display"] = df["ScreenResolution"].apply(lambda x: 1 if "Full" in x else(2 if "Ultra" in x else(3 if "Quad" in x else(4 if "Retina" in x else 0 ))))
df1.drop(['ScreenResolution'],axis=1,inplace=True)
```

```python
weight=np.array(df1['Weight'])
weight=weight.reshape(-1,1)
df1['Weight'] = scaler.fit_transform(weight)
```

```python
price=np.array(df1['Price'])
price=price.reshape(-1,1)
df1['Price'] = scaler.fit_transform(price)
```

```python
df1['Ram']=pd.factorize(df1['Ram'])[0]
df1['CPU Speed']=df1['CPU Speed'].astype('float')
df1['CPU Speed']=pd.factorize(df1['CPU Speed'])[0]
df1['CPU Vendor']=pd.factorize(df1['CPU Vendor'])[0]
df1['CPU Type']=pd.factorize(df1['CPU Type'])[0]
df1['Primary Storage']=pd.factorize(df1['Primary Storage'])[0]
df1['P Storage Type']=pd.factorize(df1['P Storage Type'])[0]
df1['Secondary Storage']=pd.factorize(df1['Secondary Storage'])[0]
df1['S Storage Type']=pd.factorize(df1['S Storage Type'])[0]
df1['GPU Vendor']=pd.factorize(df1['GPU Vendor'])[0]
df1['Vendor Type']=pd.factorize(df1['Vendor Type'])[0]
df1['OS']=pd.factorize(df1['OS'])[0]
df1['SRSize']=pd.factorize(df1['SRSize'])[0]
df1
```

```python
plt.figure(figsize=(15,10))
corr = df1.corr()
mask = np.triu(corr)
sns.heatmap(corr, annot= True, mask= mask, fmt = '.2f')
plt.title("Correlation between attributes")
plt.show()
```

# Machine Learning model

```python
from sklearn.model_selection import train_test_split
X = df1.drop('Price',axis=1) # Independet variable
y = df1['Price'] # dependent variable

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
## Importing Libraries for Models
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm 
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.model_selection import cross_val_score
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.preprocessing import PolynomialFeatures
import xgboost as xgb
```

```python
lm = LinearRegression() #LR model
lm.fit(X_train,y_train) #Fitting model
print(lm.score(X_train,y_train)) # Score
y_pred = lm.predict(X_test)
r2_score(y_test, y_pred)
```

#### XGBoost Regressor

```python
xgbr = xgb.XGBRegressor(verbosity=0)
xgbr.fit(X_train, y_train)
 
y_pred = xgbr.predict(X_test)
acc=xgbr.score(X_train,y_train)
print("Accuracy:",acc*100) # Score
print('R2 score',r2_score(y_test,y_pred))
print('MAE',mean_squared_error(y_test,y_pred))
```

```python
# Overfitting Data
#Using Cross Validation

scores = cross_val_score(xgbr, X, y, cv = 10)
scores.mean()
```
